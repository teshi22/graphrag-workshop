## Prompt Tuning（プロンプトチューニング）について

- GraphRAG には、デフォルトの汎用プロンプトのほかに、**ドメイン固有のデータ向けにプロンプトを最適化する仕組み**が用意されている。 
- 特に、**Auto Prompt Tuning** 機能を使うと、入力データをチャンク (text units) に分割し、いくつかの代表テキストを LLM に渡して解析 → そのドメイン／言語にあわせたプロンプトテンプレートを自動生成できる。
- この手法により、例えば「日本語ドキュメント」や「特定業界 (技術ブログ／決算短信など) の文章」のようなドメインでも、**より適切にエンティティ抽出や関係抽出**が行えるようになる。
- Auto チューニングは必須ではないが、**Index Run（ナレッジグラフ生成）時の結果品質向上**の観点から強く推奨されている。
- 必要に応じて、Auto で生成されたプロンプトをさらに手動で微調整 (Manual Prompt Tuning) することで、ドメイン固有の nuance や要件に対応しやすくなる。

参考：https://microsoft.github.io/graphrag/prompt_tuning/auto_prompt_tuning/
